These are essential steps in data analysis.In this stage missing values,duplicate records,incorrect formatting,derived features and data transformations are handled to improve data quality and prepare the dataset for analysis.

```python
##Handling missing values.

#Missing values are checked to identify incomplete information in the dataset.Handling missing data helps improve accuracy and prevents errors during analysis.

###Checking missing values

print("Missing Values in Dataset:")
df.isnull().sum()
```

```python
##Handling missing values in Numerical Columns.

##To handle missing values in numerical columns,the **mean (average)** value is used because it helps maintain the overall distribution of data without
#significantly affecting the dataset.

##Checking missing values in numerical columns

print(df[['Customer_Age','Store_Rating']].isnull().sum())

##Filling missing values with mean.

df['Customer_Age']=df['Customer_Age'].fillna(df['Customer_Age'].mean())
df['Store_Rating']=df['Store_Rating'].fillna(df['Store_Rating'].mean())

##Rechecking missing values.

print(df[['Customer_Age','Store_Rating']].isnull().sum())

##Explanation:
#. isnull().sum()- checks missing values
#. fillna()- fills null values
#. mean()- calculate average value
#. inplace= True- update dataset    
```

```python
##Handling missing values in Categorical columns

##Categorical columns contain labels or categories such as gender,membership status,topping type andpayment method.
##missing values incategorical columns are handled using the **mode**,which represents the most occuring value.

##Checking missing values

print(df[['Customer_Gender',
          'Membership_Status',
          'Topping_Type']].isnull().sum())

##Filling missing values

df['Customer_Gender']=df['Customer_Gender'].fillna(
    df['Customer_Gender'].mode()[0])
                       
df['Membership_Status']=df['Membership_Status'].fillna(
    df['Membership_Status'].mode()[0])

df['Topping_Type']=df['Topping_Type'].fillna(
    df['Topping_Type'].mode()[0])
      
##Rechecking missing values

print(df[['Customer_Gender',
          'Membership_Status',
          'Topping_Type']].isnull().sum())

##Explanation:
#. mode()[0]- Selects most repeated value 
      
```

```python
##Checking abnormal age values

df[df['Customer_Age']>100]

##Clean unrealistic age values
df=df[df['Customer_Age']<=100]

##Recheck maximum age
df['Customer_Age'].max()
```

```python
##Final check for missing values 

print("Missing Values in Dataset:")
print(df.isnull().sum())
```

5).Membership status v/s Total Sales (Violin plot)

The purpose of this visualization is to analyze the relationship between membership status and total sales using a violin plot.

This analysis helps understand customer spending behaviour across different membership categories and identify which membership type contributes higher sales.

Business Impact:

This visualization helps businesses understand spending patterns among different membership groups.It supports customer segmentation,targeted promotions,loyalty programs and revenue optimization stategies.

```python
#remove outliers using IQR

Q1=df['Total_Sales'].quantile(0.25)
Q3=df['Total_Sales'].quantile(0.75)
  # Q1 (First quartile):25th percentile
  # Q3 (Third quartile):75th percentile
  
IQR= Q3-Q1
 #IQR represents the middle 50% of the data

 #Calculate lower and upper limit
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR

 #Create filtered dataframe without outliers
df_filtered= df[(df['Total_Sales']>=lower_limit) &
              (df['Total_Sales']<=upper_limit)]

 #Check data set shape
print("Orginal Shape:",df.shape)
print("Filtered Shape:",df_filtered.shape)


```

```python
plt.figure(figsize=(10,6))
sns.violinplot(
    x='Membership_Status',
    y='Total_Sales',
    hue='Membership_Status',
    data=df_filtered,
    palette='coolwarm',legend=False)

plt.title('Membership Status v/s Total Sales',
          fontsize=16)
plt.xlabel('Membership Status')
plt.ylabel('Total Sales')
plt.show()
```

Observation:

The violin plot analysis revealed differences in customer spending behaviour across membership categories.Some membership groups demonstrated higher spending and stronger purchase consistency,indicating greater customer loyalty.

8.) Delivery Time Analysis by Hour of Day (HeatMap)

The heatmap visualizes the relationship between delivery time and customer ordering hours.

This analysis helps:
- Identify peak order hours.
- Analyze delivery performance.
- Improve staff allocation.
- Optimize operational efficiency.

Plotly heatmap provides interactive and visually appealing insights through color intensity and hover information.

```python
#create copy of data frame

df_filtered=df_filtered.copy()

#Rename values

df_filtered['Is_Weekend']=df_filtered['Is_Weekend'].replace({
    0:'Weekday',
    1:'Weekend'})

#create pivot table

heatmap_data=df_filtered.pivot_table(
    values='Delivery_Time_Min',
    index='Is_Weekend',
    columns='Hour_of_Day',
    aggfunc='mean')

#create heatmap

fig=px.imshow(heatmap_data,text_auto='.1f',
              color_continuous_scale='viridis',
              title='Average Delivery Time: Weekend v/s Hours of Day')

#improve layout

fig.update_layout(
    title_x=0.5,
    xaxis_title='Hour of Day',
    yaxis_title='Day Type',
width=1000,
height=400)
    
fig.show()
```

1).GroupBy Analysis. (Membership status v/s Total sales)

Groupby is a data aggregation technique used to group similar records together and calculate summary statistics such as:
- SUM
- MEAN
- COUNT
- MAXIMUM
- MINIMUM
 It helps analyze patterns and compare different categories in the dataset.

Purpose:

This analysis is performed to identify which membership category contributes the highest total sales.

Business importance:
- Identify high value customer groups.
- Improve loyalty programs.
- Create targeted marketing strategies.
- Increase customer retention.


```python
## check negative sales records.

negative_sales=df[df['Total_Sales']<0]

print(negative_sales[['Transaction_ID',
                      'Membership_Status',
                      'Total_Sales']])

Membership_analysis=df.groupby(
    'Membership_Status').agg(
    Total_Sales=('Total_Sales','sum'),
    Average_Sales=('Total_Sales','mean'),
    Maximum_Sales=('Total_Sales','max'),
    Minimum_Sales=('Total_Sales','min'),
    Total_Orders=('Transaction_ID','count'),
    Sales_Variability=('Total_Sales','std')
).reset_index()

print(Membership_analysis)
```

Observation:

- Bronze membership customers generated comparitively lower sales performance.
